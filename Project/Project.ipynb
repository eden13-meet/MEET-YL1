First Semester Final Project

The data displays a study of 10 pairs of identical twins.
The study displays each person's Corpus Collasum Surface Area, Full-Scale IQ, Head Circumference (cm), Order of Birth,Pair ID (Genotype), Sex (1=Male 2=Female), Total Surface Area (cm2), Total Brain Volume (cm3) and Body Weight (kg).

It is meant to show the correlation between physical attributes such as Brain Volume, Head Circumfrence and Gender to IQ level.



Data taken from http://lib.stat.cmu.edu/datasets/IQ_Brain_Size


```python
import csv
f = open('IQ_Brain_Size.csv', 'rb')
reader = csv.reader(f)
BV = []
IQ = []
HC = []
G  = []
for row in reader:
    print row 
    a=row[0]
    b = a.replace("\t"," ")
    c=b.split()
    e=map(float,c)
    BV.append(e[6])
    IQ.append(e[1])
    HC.append(e[2])
    G.append(e[5])
    
      
print BV
print IQ
print HC
print G
    
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Effect of Brain Volume on IQ Level')
ax.set_xlabel('Brain Volume in cm3')
ax.set_ylabel('IQ Level')
bar(BV,IQ)
savefig("BV_IQ.png")
```

This graph shows the correlation between the Brain Volume of the subjects to their IQ level. 
There is no clear correaltion between the two, as the data shows that the IQ levels range from the mid 80s to 120, with no connection to the Brain Volume of the subjects. 

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Effect of Head Circumfrence on IQ Level')
ax.set_xlabel('Head Circumfrence in cm')
ax.set_ylabel('IQ Level')
scatter(HC,IQ)
savefig("HC_IQ.png")
```

This graph shows the correlation between the Head Circumfrence of the subjects to their IQ level.
You can see that just because the circumference is larger, it does not mean that the IQ level of the subject is higher, therefore proving that their is no clear relation between the two (take that, social Darwinism).

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Effect of Gender on IQ Level')
ax.set_xlabel('Gender in Numbers (Male=1, Female=2)')
ax.set_ylabel('IQ Level')
scatter(G,IQ)
savefig("G_IQ.png")
```
