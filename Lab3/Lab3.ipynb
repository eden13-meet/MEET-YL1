```python
import csv
```

```python
f = open('cig.csv', 'rb')
reader = csv.reader(f)
tar = []
nicotine = []
for row in reader:
    a=row[0]
    b = a.replace("\t"," ")
    c=b.split()
    d=c[1:]
    e=map(float,d)
    tar.append(e[0])
    nicotine.append(e[1])
      
print tar
print nicotine
    
from pylab import *

    

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Tar & Nicotine in Brands of Cigarettes')
ax.set_xlabel('Amount of Tar')
ax.set_ylabel('Amount of Nicotine')
scatter(tar,nicotine)
savefig("cig_scatter_labels.png")

```
